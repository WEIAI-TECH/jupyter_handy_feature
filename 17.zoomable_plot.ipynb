```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt


# define an array x of 32 uniformly spaced floating point
# values from 0 to 4Ï€
x = np.linspace(0,4*np.pi,32)

# define y as the array containing the sine values from x
y = np.sin(x)

# create the plot
fig,ax = plt.subplots()
ax.plot(x, y)
```
